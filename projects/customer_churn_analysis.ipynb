```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("customer churn dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print(df.columns)
```

Churn Distribution

```python
sns.countplot(x="Churn", data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

Tenure vs Churn

```python
sns.boxplot(x="Churn", y="Tenure", data=df)
plt.title("Tenure vs Churn")
plt.show()
```

Contract Length vs Churn

```python
sns.countplot(x="Contract Length", hue="Churn", data=df)
plt.title("Contract Length vs Churn")
plt.show()
```

Usage Frequency vs Churn

```python
sns.boxplot(x="Churn", y="Usage Frequency", data=df)
plt.title("Usage Frequency vs Churn")
plt.show()
```

Support Calls vs Churn

```python
sns.boxplot(x="Churn", y="Support Calls", data=df)
plt.title("Support Calls vs Churn")
plt.show()
```

Total Spend vs Churn

```python
sns.boxplot(x="Churn", y="Total Spend", data=df)
plt.title("Total Spend vs Churn")
plt.show()
```

Contract Length vs Churn

```python
sns.countplot(x="Contract Length", hue="Churn", data=df)
plt.title("Contract Length vs Churn")
plt.show()
```

1. Customers with shorter tenure tend to churn more frequently.

2. Customers with lower usage frequency show higher churn probability.

3. A higher number of support calls is associated with increased churn risk.

4. Customers with shorter contract lengths are more likely to churn.

5. Customers with lower total spending are more likely to leave the service.
