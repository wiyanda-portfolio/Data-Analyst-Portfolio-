```python
# load dataset
import pandas as pd

df = pd.read_csv("sales_dataset.csv", sep=";")
df.head()
```

```python
print(df.columns)
```

```python
revenue_product = df.groupby("product")["total_sales"].sum().sort_values(ascending=False)

plt.figure()
revenue_product.plot(kind="bar")
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
city_sales = df.groupby("city")["total_sales"].sum().sort_values(ascending=False)

plt.figure()
city_sales.plot(kind="bar")
plt.title("Sales by City")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.show()
```
