# Sales Data Analysis
This project analyzes sales data to find insights such as top products, top cities, and most purchased products.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('sales_dataset.csv', sep=';')
df.head()
```

```python
df.groupby('product')['total_sales'].sum().sort_values().plot(kind='barh')
plt.title('Total Revenue by Product')
plt.show()
```

```python
df.groupby('city')['total_sales'].sum().sort_values().plot(kind='barh')
plt.title('Total Sales by City')
plt.show()
```

```python
df.groupby('product')['quantity'].sum().sort_values().plot(kind='barh')
plt.title('Most Purchased Product')
plt.show()
```
